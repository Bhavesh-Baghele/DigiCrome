```python
plt.figure(figsize=(12,6))
days=[1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.stackplot(days,sleeping,eating,working,playing,
              labels=('Sleeping','Eating','Working','Playing'),
              colors=['m','c','r','g'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Day wise activity time spent - pattern')
plt.legend()
plt.show()
```

```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']

plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=0,# default value , start at +ve x axis,if 90 rotate counter
        shadow= True,
        explode=(0.1,0,0,0),  #creat gap
        autopct='%1.4f%%')  # show u the percentage
plt.title('Day wise activity time spent - pattern')
plt.show()
```
